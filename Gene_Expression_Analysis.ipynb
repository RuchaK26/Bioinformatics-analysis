```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
control = df[df["Outcome"] == 0]
disease = df[df["Outcome"] == 1]
```

```python
from scipy.stats import ttest_ind
ttest_ind(control["Glucose"], disease["Glucose"])
```

We performed an independent t-test to compare this feature between individuals with Outcome = 0 (non-diabetic) and Outcome = 1 (diabetic). The p-value helps determine whether the observed difference is statistically meaningful. A p-value below 0.05 indicates a statistically significant difference between the two groups.

The independent t-test shows a very large negative t-statistic and an extremely small p-value (p ≈ 8.9 × 10⁻⁴³). Since p < 0.05, the difference in Glucose levels between Outcome = 0 (non-diabetic) and Outcome = 1 (diabetic) groups is statistically highly significant. This means glucose levels are meaningfully higher in diabetic individuals compared to non-diabetic individuals.

```python
sns.boxplot(x="Outcome", y="Glucose", data=df)
plt.title("Glucose Levels in Control vs Disease")
plt.show()
```

```python
sns.heatmap(df.corr(), cmap="viridis")
plt.title("Feature Correlation Heatmap")
plt.show()
```
